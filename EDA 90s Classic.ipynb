```python
##Converting this into a dict
df_dict=df.groupby('Time_Signature')['Track'].count().to_dict()
```

```python
df_dict
```

```python
##Extarcting the Time Signatures
X=[i for i in df_dict.keys()]
```

```python
X
```

```python
##Extarcting the Counts
Y=[i for i in df_dict.values()]
```

```python
Y
```

```python
##Creating a Pie Chart
sns.set_theme()
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.pie(Y,labels=X)
```

```python
bar_dict=df.groupby('Year')['Track'].count().to_dict()
X=[i for i in bar_dict.keys()]
Y=[i for i in bar_dict.values()]
```

```python
##Creating 4 Charts using subcharts
fig,axes=plt.subplots(2,2,figsize=(10,10))
sns.barplot(x=X,y=Y,ax=axes[0,0])
axes[0,0].set_title('Count of Songs per Year')
sns.histplot(data=df,x='Loudness',hue='Year',ax=axes[0,1],kde=True)
axes[0,1].set_title('Histogram of Loudness year Wise')
sns.scatterplot(data=df,x='Popularity',y='Loudness',hue='Time_Signature',ax=axes[1,0])
axes[1,0].set_title('Scatterplot of Loudness and popularity')
sns.distplot(df['Tempo'],ax=axes[1,1])
```
